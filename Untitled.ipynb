```python
import requests
import pandas as pd
```

```python
res = requests.get(url = "https://api.weather.gov/gridpoints/LWX/89,70/forecast/hourly")
```

```python
j = res.json()
```

```python
periods = j["properties"]["periods"]
```

```python
df = pd.DataFrame(periods)
```

```python
df = pd.DataFrame(j["properties"]["periods"])
df.to_csv("hourly_weather.csv", index=False)
df.head()
```

```python
type(j), list(j.keys())           
props = j.get('properties', {})
print("properties keys:", list(props.keys()))

# check possible containers where hourly entries might be:
for candidate in ['periods','timeseries','periods', 'items', 'features']:
    if candidate in props:
        print(f"Found '{candidate}' with length:", len(props[candidate]) if hasattr(props[candidate], '__len__') else "unknown")
```

```python
container_key = 'periods'   
entries = j['properties'][container_key]

# quick sanity
len(entries)
```

```python
df = pd.json_normalize(entries)
df.shape, list(df.columns)[:30]   # show shape and first 30 columns
df.head()
```

```python
cols_wanted = ['temperature', 'temperatureUnit', 'windSpeed_mph', 'windDirection', 'shortForecast']
cols_wanted = [c for c in cols_wanted if c in df.columns]  
df_small = df[cols_wanted].copy()

    
    
df_small.head(10)
```

```python
csv_filename = 'hourly_forecast.csv'
df_small.to_csv(csv_filename, index=False)
print("Saved:", csv_filename)
```
